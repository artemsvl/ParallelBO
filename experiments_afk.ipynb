```python
from run_experiments import run_experiment
N_RUNS = 30
N_ITERATIONS = 20
dimensions = [4, 8, 16]
q = 50

exp_results = {'qlogei': {}, 'async': {}}

for dim in dimensions:
    n_init = 3 * q
    print(f"\nRunning Experiment 3: q={q}, d={dim}")
    print("="*70)
    
    exp_results['qlogei'][dim] = run_experiment(
        strategy="qlogei",
        dim=dim,
        batch_size=q,
        n_init=n_init,
        n_iterations=N_ITERATIONS,
        n_runs=N_RUNS
    )
    
    exp_results['async'][dim] = run_experiment(
        strategy="async_simulation",
        dim=dim,
        batch_size=q,
        n_init=n_init,
        n_iterations=N_ITERATIONS,
        n_runs=N_RUNS
    )
```
