# Parallel Bayesian Optimization Experiments

This notebook compares qLogEI and AsyncSimulation strategies across different dimensions and batch sizes.

```python
import json
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from run_experiments import run_experiment

plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## Experiment 1: q=1 Analysis

**Goal:** Compare qLogEI and AsyncSimulation strategies with batch size q=1.

**Hypothesis:** With q=1, both strategies should behave similarly since there's no parallelization.

**Parameters:**
- Dimension: d=1
- Batch size: q=1
- Number of runs: N_RUNS=3
- Iterations: 20

```python
N_RUNS = 3
N_ITERATIONS = 20
dim = 1
q = 1
n_init = 3 * dim

print("Running Experiment 1: q=1, d=1")
print("="*70)

exp1_qlogei = run_experiment(
    strategy="qlogei",
    dim=dim,
    batch_size=q,
    n_init=n_init,
    n_iterations=N_ITERATIONS,
    n_runs=N_RUNS
)

exp1_async = run_experiment(
    strategy="async_simulation",
    dim=dim,
    batch_size=q,
    n_init=n_init,
    n_iterations=N_ITERATIONS,
    n_runs=N_RUNS
)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

qlogei_dir = Path(f"data/qLogEI/dim={dim}/q={q}/{exp1_qlogei['timestamp']}")
async_dir = Path(f"data/AsyncSimulation/dim={dim}/q={q}/{exp1_async['timestamp']}")

colors = ['#FF6B6B', '#4ECDC4', '#FFD93D']

for run_idx in range(N_RUNS):
    with open(qlogei_dir / f"run_{run_idx:03d}.json") as f:
        qlogei_data = json.load(f)
    with open(async_dir / f"run_{run_idx:03d}.json") as f:
        async_data = json.load(f)
    
    y_qlogei = np.array(qlogei_data['y_all'])
    y_async = np.array(async_data['y_all'])
    
    best_qlogei = np.maximum.accumulate(y_qlogei)
    best_async = np.maximum.accumulate(y_async)
    
    axes[0].plot(best_qlogei, alpha=0.85, linewidth=2.5, color=colors[run_idx], label=f'Run {run_idx+1}')
    axes[1].plot(best_async, alpha=0.85, linewidth=2.5, color=colors[run_idx], label=f'Run {run_idx+1}')

axes[0].set_title('qLogEI (q=1, d=1)', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Number of Evaluations', fontsize=12)
axes[0].set_ylabel('Best Value Found', fontsize=12)
axes[0].legend(fontsize=11)
axes[0].grid(True, alpha=0.3)

axes[1].set_title('AsyncSimulation (q=1, d=1)', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Number of Evaluations', fontsize=12)
axes[1].set_ylabel('Best Value Found', fontsize=12)
axes[1].legend(fontsize=11)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "="*70)
print("Experiment 1 Results:")
print(f"qLogEI       - Mean: {exp1_qlogei['mean_best']:.4f} ± {exp1_qlogei['std_best']:.4f}")
print(f"AsyncSim     - Mean: {exp1_async['mean_best']:.4f} ± {exp1_async['std_best']:.4f}")
print("="*70)
```

## Experiment 2: q=2 with Increasing Dimensions

**Goal:** Analyze how both strategies perform with q=2 as dimensionality increases.

**Parameters:**
- Dimensions: d=[1, 2, 4, 8, 16]
- Batch size: q=2
- Number of runs: N_RUNS=3
- Iterations: 20

```python
N_RUNS = 3
N_ITERATIONS = 20
dimensions = [1, 2, 4, 8, 16]
q = 2

exp2_results = {'qlogei': {}, 'async': {}}

for dim in dimensions:
    n_init = 3 * dim
    print(f"\nRunning Experiment 2: q={q}, d={dim}")
    print("="*70)
    
    exp2_results['qlogei'][dim] = run_experiment(
        strategy="qlogei",
        dim=dim,
        batch_size=q,
        n_init=n_init,
        n_iterations=N_ITERATIONS,
        n_runs=N_RUNS
    )
    
    exp2_results['async'][dim] = run_experiment(
        strategy="async_simulation",
        dim=dim,
        batch_size=q,
        n_init=n_init,
        n_iterations=N_ITERATIONS,
        n_runs=N_RUNS
    )
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for idx, dim in enumerate(dimensions):
    if idx >= 6:
        break
    
    qlogei_dir = Path(f"data/qLogEI/dim={dim}/q={q}/{exp2_results['qlogei'][dim]['timestamp']}")
    async_dir = Path(f"data/AsyncSimulation/dim={dim}/q={q}/{exp2_results['async'][dim]['timestamp']}")
    
    for run_idx in range(N_RUNS):
        with open(qlogei_dir / f"run_{run_idx:03d}.json") as f:
            qlogei_data = json.load(f)
        with open(async_dir / f"run_{run_idx:03d}.json") as f:
            async_data = json.load(f)
        
        y_qlogei = np.array(qlogei_data['y_all'])
        y_async = np.array(async_data['y_all'])
        
        best_qlogei = np.maximum.accumulate(y_qlogei)
        best_async = np.maximum.accumulate(y_async)
        
        axes[idx].plot(best_qlogei, alpha=0.6, label=f'qLogEI-{run_idx+1}', linestyle='-')
        axes[idx].plot(best_async, alpha=0.6, label=f'Async-{run_idx+1}', linestyle='--')
    
    axes[idx].set_title(f'd={dim}, q={q}')
    axes[idx].set_xlabel('Number of Evaluations')
    axes[idx].set_ylabel('Best Value Found')
    axes[idx].legend(fontsize=8)
    axes[idx].grid(True)

if len(dimensions) == 5:
    axes[5].axis('off')

plt.tight_layout()
plt.show()

fig, ax = plt.subplots(1, 1, figsize=(10, 6))

qlogei_means = [exp2_results['qlogei'][d]['mean_best'] for d in dimensions]
qlogei_stds = [exp2_results['qlogei'][d]['std_best'] for d in dimensions]
async_means = [exp2_results['async'][d]['mean_best'] for d in dimensions]
async_stds = [exp2_results['async'][d]['std_best'] for d in dimensions]

ax.errorbar(dimensions, qlogei_means, yerr=qlogei_stds, marker='o', label='qLogEI', capsize=5, linewidth=2)
ax.errorbar(dimensions, async_means, yerr=async_stds, marker='s', label='AsyncSimulation', capsize=5, linewidth=2)

ax.set_xlabel('Dimension (d)', fontsize=12)
ax.set_ylabel('Mean Best Value', fontsize=12)
ax.set_title('Performance vs Dimension (q=2)', fontsize=14)
ax.legend(fontsize=12)
ax.grid(True)
ax.set_xscale('log', base=2)

plt.tight_layout()
plt.show()

print("\n" + "="*70)
print("Experiment 2 Results: q=2 with varying dimensions")
print("="*70)
print(f"{'Dim':<6} {'qLogEI Mean':<15} {'AsyncSim Mean':<15} {'Difference':<15}")
print("-"*70)
for dim in dimensions:
    qlogei_mean = exp2_results['qlogei'][dim]['mean_best']
    async_mean = exp2_results['async'][dim]['mean_best']
    diff = async_mean - qlogei_mean
    print(f"{dim:<6} {qlogei_mean:<15.4f} {async_mean:<15.4f} {diff:+15.4f}")
print("="*70)
```
